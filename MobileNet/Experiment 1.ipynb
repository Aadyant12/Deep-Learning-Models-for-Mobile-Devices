**THIS IS THE MAIN MOBILENET NETWORK** 

```python
from tensorflow.keras.datasets import cifar10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
# example of loading the MobileNet model
from tensorflow.keras.applications.mobilenet import MobileNet
# load model
model = MobileNet(input_shape=(32,32,3), include_top=False, weights=None)
# summarize the model
model.summary()
```

```python
len(model.layers)
```

```python
# add new classifier layers for cifar10 classification
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten

flat = Flatten()(model.layers[-1].output)
class1 = Dense(512, activation='relu')(flat)
class2 = Dense(256, activation='relu')(class1)
class3 = Dense(128, activation='relu')(class2)

prediction = Dense(10, activation='softmax')(class3)

model = Model(inputs=model.inputs, outputs=prediction)
model.summary()
```

```python
model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
```

```python
model.fit(
    x=x_train,
    y=y_train,
    epochs=20,
    verbose=1,
    validation_data=(x_test, y_test),
    batch_size=128
)
```

```python
i = 0
exits = []
for layer in model.layers:
  if 'pad' in layer.name:
    print("Exit layer number: ", i, ", Output shape of exit layer: ", layer.output_shape)
    exits.append(i)
  i += 1
```

```python
i = 1
for exit in exits:
  print(exit)
  exit_layer = model.layers[exit]
  Model(inputs = model.inputs, outputs = exit_layer.output).save_weights(f"/MobileNet/Main_till_exit{i}_weights.h5")
  i += 1
```

```python
import time
```

```python
start = time.time()
history = model.fit(
    x=x_train,
    y=y_train,
    epochs=100,
    verbose=1,
    validation_data=(x_test, y_test),
    batch_size=128
)
end = time.time()
```
