### **HERE EXIT NETWORKS HAVE BEEN ADDED IN MOBILENET NETWORK**

```python
from tensorflow.keras.datasets import cifar10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
from tensorflow.keras.applications import EfficientNetB0
model = EfficientNetB0(include_top=False, weights=None, input_shape=(32,32,3))
# summarize the model
model.summary()
```

```python
len(model.layers)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
```

```python
exits = [45, 74, 103, 118, 146, 161, 190, 205]
```

```python
for model in exit_models:
  model.compile(
          optimizer='adam',
          loss='sparse_categorical_crossentropy',
          metrics=['accuracy']
      )
```

```python
import time
```

```python
i = 1
accuracies_time = []
for model in exit_models:
  start = time.time()
  history = model.fit(
      x=x_train,
      y=y_train,
      epochs=100,
      verbose=1,
      validation_data=(x_test, y_test),
      batch_size=128
  )
  end = time.time()
  Model(inputs = model.inputs, outputs = model.layers[exits[i-1]].output).save_weights(f"/EffNet/EarlyExit_till_exit{i}_weights.h5")
  print("Saved!")
  accuracies_time.append(((history.history.get('val_accuracy')[len(history.history.get('val_accuracy')) - 1]), (end-start)))
  print(f"Accuracy of model {i}: ", history.history.get('val_accuracy')[len(history.history.get('val_accuracy')) - 1])
  i += 1
```
